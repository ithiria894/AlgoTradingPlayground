```python
#calculate the 最大连续亏损次数 和 最大连续盈利次数 using Profit % column, if the profit is positive, it is a win, otherwise it is a loss
# Create a list to store the consecutive win counts
consecutive_wins = []

# Create a list to store the consecutive loss counts
consecutive_losses = []

# Create a variable to store the current consecutive win count
current_consecutive_win = 0

# Create a variable to store the current consecutive loss count
current_consecutive_loss = 0

# Iterate over the profit percentages

for profit in df['Profit %']:
    # If the profit is positive, it is a win
    if profit > 0:
        # Increment the current consecutive win count
        current_consecutive_win += 1
        # Reset the current consecutive loss count
        current_consecutive_loss = 0
    # If the profit is negative, it is a loss
    elif profit < 0:
        # Increment the current consecutive loss count
        current_consecutive_loss += 1
        # Reset the current consecutive win count
        current_consecutive_win = 0
    # If the profit is zero, it is neither a win nor a loss
    else:
        # Reset the current consecutive win count
        current_consecutive_win = 0
        # Reset the current consecutive loss count
        current_consecutive_loss = 0
    # Append the current consecutive win count to the list of consecutive wins
    consecutive_wins.append(current_consecutive_win)
    # Append the current consecutive loss count to the list of consecutive losses
    consecutive_losses.append(current_consecutive_loss)

# Calculate the maximum consecutive win count
max_consecutive_wins = max(consecutive_wins)

# Calculate the maximum consecutive loss count
max_consecutive_losses = max(consecutive_losses)

# Print the result
print('Maximum Consecutive Wins:', max_consecutive_wins)
print('Maximum Consecutive Losses:', max_consecutive_losses)
```

```python
#no compounding
import random

def generate_trade_list(trade_count, win_rate, max_consecutive_wins, max_consecutive_losses):
    win_count = int(trade_count * win_rate)
    loss_count = trade_count - win_count

    # 生成初始的布尔列表，大约70%是1，大约30%是0
    trade_list = [1] * win_count + [0] * loss_count
    random.shuffle(trade_list)

    # 确保连续出现9个1之后下一个一定是0
    consecutive_wins = 0
    for i in range(len(trade_list)):
        if trade_list[i] == 1:
            consecutive_wins += 1
            if consecutive_wins >= max_consecutive_wins:
                if i+1 < len(trade_list) and trade_list[i+1] == 1:
                    trade_list[i+1] = 0
                    consecutive_wins = 0
        else:
            consecutive_wins = 0

    # 确保连续出现3个0之后下一个一定是1
    consecutive_losses = 0
    for i in range(len(trade_list)):
        if trade_list[i] == 0:
            consecutive_losses += 1
            if consecutive_losses >= max_consecutive_losses:
                if i+1 < len(trade_list) and trade_list[i+1] == 0:
                    trade_list[i+1] = 1
                    consecutive_losses = 0
        else:
            consecutive_losses = 0
    
    #calculate the percentage of wins in trade_list
    win_rate = trade_list.count(1)/len(trade_list)
    print("Win Rate:", win_rate)

    return trade_list

def calculate_profit(trade_list, initial_capital, leverage):
    profit = 0

    for trade in trade_list:
        if trade == 1:
            win_percent = random.uniform(0.0032, 0.023)
            profit += initial_capital * leverage * win_percent  # 每次赢得最多2.3%
        else:
            loss_percent = random.uniform(0.0062, 0.019)
            profit -= initial_capital * leverage * loss_percent  # 每次输最多1.9%

    return profit

# 示例用法
initial_capital = 1000
leverage = 1
trade_count = 100
win_rate = 0.7
max_consecutive_wins = 15
max_consecutive_losses = 9

trade_list = generate_trade_list(trade_count, win_rate, max_consecutive_wins, max_consecutive_losses)
final_profit = calculate_profit(trade_list, initial_capital, leverage)
print("Final Profit:", final_profit)
```

```python
#with compounding

def generate_trade_list(trade_count, win_rate, max_consecutive_wins, max_consecutive_losses):
    win_count = int(trade_count * win_rate)
    loss_count = trade_count - win_count

    # 生成初始的布尔列表，大约70%是1，大约30%是0
    trade_list = [1] * win_count + [0] * loss_count
    random.shuffle(trade_list)

    # 确保连续出现9个1之后下一个一定是0
    consecutive_wins = 0
    for i in range(len(trade_list)):
        if trade_list[i] == 1:
            consecutive_wins += 1
            if consecutive_wins >= max_consecutive_wins:
                if i+1 < len(trade_list) and trade_list[i+1] == 1:
                    trade_list[i+1] = 0
                    consecutive_wins = 0
        else:
            consecutive_wins = 0

    # 确保连续出现3个0之后下一个一定是1
    consecutive_losses = 0
    for i in range(len(trade_list)):
        if trade_list[i] == 0:
            consecutive_losses += 1
            if consecutive_losses >= max_consecutive_losses:
                if i+1 < len(trade_list) and trade_list[i+1] == 0:
                    trade_list[i+1] = 1
                    consecutive_losses = 0
        else:
            consecutive_losses = 0

    return trade_list

def calculate_profit(trade_list, initial_capital, leverage, compounding):
    profit = 0
    capital = initial_capital

    for trade in trade_list:
        if trade == 1:
            win_percent = random.uniform(0.0032, 0.0338)
            # print(win_percent)
            profit += initial_capital * leverage * win_percent  # 每次赢得最多2.3%
        else:
            # loss_percent = random.uniform(0.0062, 0.019)
            loss_percent = random.uniform(0.0062, 0.0243)
            profit -= initial_capital * leverage * loss_percent  # 每次输最多1.9%

        if compounding:  # 考虑是否使用复利
            capital += profit

    return profit

# 示例用法
initial_capital = 100
leverage = 40
trade_count = 100
win_rate = 0.7
max_consecutive_wins = 15
max_consecutive_losses = 9
compounding = False

trade_list = generate_trade_list(trade_count, win_rate, max_consecutive_wins, max_consecutive_losses)
#calculate the percentage of wins in trade_list
win_rate = trade_list.count(1)/len(trade_list)
print("Win Rate:", win_rate)
#calculate the 连续亏损次数 和 连续盈利次数
consecutive_wins = 0
consecutive_losses = 0
current_consecutive_win = 0
current_consecutive_loss = 0
for trade in trade_list:
    if trade == 1:
        current_consecutive_win += 1
        current_consecutive_loss = 0
    elif trade == 0:
        current_consecutive_loss += 1
        current_consecutive_win = 0
    if current_consecutive_win > consecutive_wins:
        consecutive_wins = current_consecutive_win
    if current_consecutive_loss > consecutive_losses:
        consecutive_losses = current_consecutive_loss
print("Consecutive Wins:", consecutive_wins)
print("Consecutive Losses:", consecutive_losses)

final_profit = calculate_profit(trade_list, initial_capital, leverage, compounding)
print("Final Profit:", final_profit)
```

```python
import matplotlib.pyplot as plt

# Group the dataframe by month and count the number of positive and negative profits for each month
df['Year-Month'] = df.index.to_period('M')
profit_counts = df.groupby(['Year-Month', df['Profit %'] > 0]).size().unstack().fillna(0)
profit_counts.columns = ['Negative Profit', 'Positive Profit']  # Rename the columns for clarity

# Plot the result using a bar chart
plt.figure(figsize=(12, 6))

bar_width = 0.35
index = range(len(profit_counts.index))
plt.bar(index, profit_counts['Positive Profit'], bar_width, label='Positive Profit')
plt.bar([i + bar_width for i in index], profit_counts['Negative Profit'], bar_width, label='Negative Profit')

plt.title('Profit Counts by Year-Month')
plt.xlabel('Year-Month')
plt.ylabel('Count')
plt.xticks([i + bar_width/2 for i in index], profit_counts.index, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
