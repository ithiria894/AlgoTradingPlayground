```python
import ccxt
import pandas as pd
import datetime
from datetime import timedelta
from datetime import datetime
import pytz
# exchange = ccxt.binance()
# symbol = 'DOGE/USDT'
# timeframe = '1m'
# from_date = '2023-04-17'
# to_date = '2023-04-25'

# ohlcv = exchange.fetch_ohlcv(symbol, timeframe, exchange.parse8601(from_date), exchange.parse8601(to_date))

# df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
# df['dateandTime'] = pd.to_datetime(df['timestamp'], unit='ms')
# # df['average_price'] = (df['open'] + df['high'] + df['low'] + df['close']) / 4
# df = df[['dateandTime', 'open', 'high', 'low', 'close', 'volume']]
# import ccxt
# import pandas as pd
# from datetime import datetime
# from datetime import timedelta

exchange = ccxt.binance()
symbol = 'BTC/USDT'
timeframe = '1m'
from_date = '2020-12-15'
to_date = '2020-12-20'

batch_size = 500
data = []
start_time = datetime.strptime(from_date, '%Y-%m-%d')
end_time = datetime.strptime(to_date, '%Y-%m-%d')

print(f'start_time: {start_time.timestamp()}')
print(f'end_time: {end_time.timestamp()}')

while start_time < end_time:
    ohlcv = exchange.fetch_ohlcv(symbol, timeframe, int(start_time.timestamp()*1000), limit=batch_size)
    data.extend(ohlcv)
    start_time = datetime.fromtimestamp(ohlcv[-1][0]/1000) + timedelta(minutes=1)

df = pd.DataFrame(data, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])
df['dateandTime'] = pd.to_datetime(df['timestamp'], unit='ms', utc=True)
df = df[['dateandTime', 'open', 'high', 'low', 'close', 'volume']]
df
```

```python
from datetime import datetime, timedelta

# Dates of Bitcoin halving events
halving_dates = [
    datetime(2012, 11, 28),
    datetime(2016, 7, 9),
    datetime(2020, 5, 11),
    datetime(2024, 4, 18)  # Future halving date
]

# Dates of subsequent all-time highs
all_time_high_dates = [
    datetime(2013, 12, 4),
    datetime(2017, 12, 17),
    datetime(2021, 3, 13)
]

# Calculate durations between halving and subsequent all-time high in months
durations = [(all_time_high_dates[i] - halving_dates[i]).days / 30 for i in range(len(halving_dates) - 1)]

# Calculate average duration
average_duration = sum(durations) / len(durations)

# Predict next all-time high date
future_halving_date = halving_dates[-1]
predicted_all_time_high_date = future_halving_date + timedelta(days=int(average_duration * 30))

print("Duration between each halving and subsequent all-time high (in months):")
for i, duration in enumerate(durations, start=1):
    print(f"Halfing {i} to All-Time High: {duration:.2f} months")

print("\nAverage Duration:", f"{average_duration:.2f}", "months")

print("\nPredicted All-Time High Date after the 4th halving:", predicted_all_time_high_date.strftime("%Y-%m-%d"))
```
